Tải vài nạp dữ liệu

```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

Lấy dữ liệu (20 từ đầu tiên):

```python
for index, word in enumerate(wv.index2word):
    if index == 20:
        break
    print(f"từ vựng #{index}/{len(wv.index2word)} là {word}")
```

Có thể lấy được vectơ cho các thuật ngữ mà mô hình đã đào tạo

```python
vec_queen = wv['queen']
```

Điểm yếu của Word2Vec là không thể lấy vectơ từ không quen thuộc

```python
try:
    vec_cameroon = wv['cameroon']
except KeyError:
    print("Từ 'cameroon' không tồn tại bên trong mô hình!")
```

Lấy top 10 từ gần với từ “car”:

```python
print(wv.most_similar(positive=['car'], topn = 10))
```

Kiểm tra: Các từ có mức độ giống nhau giảm dần

```python
pairs = [
    ('car', 'minivan'),   # 'minivan' là một loại của 'car'
    ('car', 'bicycle'),   # vẫn là xe có bánh
    ('car', 'airplane'),  # không có bánh, nhưng nó là vẫn là một phương tiện di chuyển
    ('car', 'cereal'),    # và tiếp tục giảm dần...
    ('car', 'communism'),
]
for w1, w2 in pairs:
    print('%r\t%r\t%.2f' % (w1, w2, wv.similarity(w1, w2)))
```

Kiểm tra và in ra từ khác biệt

```python
print(wv.doesnt_match(['fire', 'water', 'land', 'sea', 'air', 'car']))
```
